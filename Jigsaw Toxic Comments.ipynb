```python
#Regular Importing
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
# Needed Stuff
import gensim
import nltk

from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
nltk.download('punkt')

import gensim.downloader as api
from gensim.models import Word2Vec
```

```python
comments = pd.read_csv("comments_to_score.csv")
comments
```

```python
# Tokenize the string into words
comments_text = comments["text"]
tokenizer = RegexpTokenizer(r'\w+')
comments_list = [tokenizer.tokenize(comment) for comment in comments_text]
comments_list
```

```python
model = Word2Vec(comments_list, min_count=1,workers=5, window =3, sg = 1)
```
